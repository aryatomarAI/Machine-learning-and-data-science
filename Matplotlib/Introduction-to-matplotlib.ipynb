## pyplot vs matplotlib oo method

* When plotting something quikly , it's okay to use the pyplot method
* When plotting something more advanced, use the oo method of matplotlib

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease['age']>50]
```

```python
over_50.head()
```

```python
# pyplot method
```

```python
over_50.plot(kind='scatter',x='age',y='chol',c='target');
```

```python
# combined pyplot method and oo method
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
# ax.set_xlim([45,100])
```

## OO Method from scratch

```python
fig,ax=plt.subplots(figsize=(10,6))
# Plot The data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50["target"])
```

```python
# customize the plot 
```

```python
fig,ax=plt.subplots(figsize=(10,6))
# Plot The data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50["target"])
# customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",xlabel='Age', ylabel='Cholestrol');
```

```python
# Add a legend 
```

```python
fig,ax=plt.subplots(figsize=(10,6))
# Plot The data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50["target"])
# customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",xlabel='Age', ylabel='Cholestrol')
# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")
# Add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
# Plotting Subplots
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)  # Both ploths share same x axis
# For Ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax0.set(title="Heart Disease and cholestrol Levels",ylabel='cholestrol')
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50['chol'].mean(),linestyle="--")
# For Ax1
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(title="Heart Disease and maximum heart rate",xlabel="Age",ylabel="Maximum Heart Rate")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50['thalach'].mean(),linestyle="--")
# Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```

```python
# Customizing heart Disease data 
plt.style.use('seaborn-whitegrid')
fig,ax=plt.subplots(figsize=(10,6))
# Plot The data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50["target"],cmap='winter') # use cmap to change colour
# customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",xlabel='Age', ylabel='Cholestrol')
# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")
# Add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
# Customizing x and y axis limitations
plt.style.use('seaborn-whitegrid')
fig, (ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)  # Both ploths share same x axis
# For Ax0
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
ax0.set(title="Heart Disease and cholestrol Levels",ylabel='cholestrol')
# Change the x axis limits
ax0.set_xlim([50,80])
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50['chol'].mean(),linestyle="--",color='red')
# For Ax1
scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter")
ax1.set(title="Heart Disease and maximum heart rate",xlabel="Age",ylabel="Maximum Heart Rate")
# Change limitation of y axis
ax1.set_ylim([60,200])
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50['thalach'].mean(),linestyle="--",color='red')
# Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```
