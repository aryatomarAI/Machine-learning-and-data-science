```python
# Make a plot from dictionary
```

```python
nut_butter_prices={"Almond butter":10,
                  "Peanut butter":8,
                  "Cashew butter":12}
```

```python
fig, ax=plt.subplots()              # Bar Plot
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",ylabel="Prices (Rs)", xlabel="Butter Variety");
```

```python
# Histogram and subplots
```

```python
fig, ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax.set(title="Dan's Nut Butter Store", ylabel="Butter Variety",xlabel="Prices (Rs)");
```

```python
# Make some Data
```

```python
x=np.random.randn(1000) # standard normal distribution
fig,ax=plt.subplots()   
ax.hist(x);
```

```python
# Two options of subpLots
```

```python
# Subplot option 1
```

```python
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
# Plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

```python
car_sales=pd.read_csv(r'C:\Users\DELL\Desktop\Machine-learning-and-data-science\cars-sales.csv')
```

```python
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[$,]','').astype(float)
```

```python
car_sales
```

```python
car_sales["Sale Date"]=pd.date_range("2/3/2021",periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total Sale"]=car_sales["Price"].astype(int).cumsum()
```

```python
car_sales
```

```python
# Plotting graph between total sales on corresponding days
```

```python
car_sales.plot(x="Sale Date",y="Total Sale");
```

```python
# Changing 'Price' Data type to int as currently it is of string type
car_sales["Price"]=car_sales["Price"].astype(int)

car_sales.plot(x='Odometer (KM)',y="Price",kind="scatter");
```
