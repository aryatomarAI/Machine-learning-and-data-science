## Add or Create Data with existing Data

```python
# Add new column to your data with the help of series method
```

```python
seats_column=pd.Series([5,5,5,5,5])
car_sales["Seats"]=seats_column
```

```python
car_sales
```

```python
# adding column with the help of lists
# But length of list should be equal to the length of your csv file 
```

```python
fuel_economy=[4,5,5.6,7.8,9.0,11,13.4,11,9.8,12]
car_sales["Fuel per 100 Km"]=fuel_economy
```

```python
car_sales
```

```python
# Create a new column from another column
```

```python
car_sales["Total fuel used"]=car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100 Km"]
```

```python
car_sales
```

```python
 # Removing columns
```

```python
car_sales.drop("Total fuel used",axis=1,inplace=True)
```

```python
# Shuffling our data 
```

```python
car_sales_shuffled=car_sales.sample(frac=1)
car_sales_shuffled
```

```python
car_sales_shuffled.sample(frac=0.2)    # 20% of our data
```

```python
# Reset Index
```

```python
car_sales_shuffled.reset_index(inplace=True,drop=True)  # We are using drop attribute to avoid additional indexing
car_sales_shuffled
```
